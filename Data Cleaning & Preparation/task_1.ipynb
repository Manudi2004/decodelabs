```python
##Section 3:  Handle Missing Values (CouponCode)
#Issue:CouponCode has 309 null values (25.75%)
#Decision: Impute with 'NONE' — not delete rows.
#Justification:A missing coupon code means no coupon was used, not that the record is invalid.
#Deleting these rows would reduce statistical power and lose valid order data.
```

```python
# Before
print("Before fix:")
print(f"  Null values in CouponCode: {df['CouponCode'].isnull().sum()}")
print(f"  Sample nulls:\n{df[df['CouponCode'].isnull()][['OrderID', 'CouponCode']].head(5)}")
```

```python
# Apply fix
df['CouponCode'] = df['CouponCode'].fillna('NONE')

# After (verify)
print("After fix:")
print(f"  Null values in CouponCode: {df['CouponCode'].isnull().sum()}")
print(f"  Unique values: {df['CouponCode'].unique()}")
```

```python
## Section 4: Fix Floating Point Precision (UnitPrice, TotalPrice)
#Issue: UnitPrice and TotalPrice contain floating point errors (e.g. 550.6799999999999)
#Decision:Round both columns to 2 decimal places.

#Justification:Monetary values must always be represented with exactly 2 decimal places
#to ensure accuracy in financial calculations and reporting.
```

```python
# Before
print("Before fix:")
print("UnitPrice samples with precision issues:")
issues = df[df['UnitPrice'] != df['UnitPrice'].round(2)]['UnitPrice']
print(issues.head(10).tolist())
print(f"\nTotal UnitPrice precision issues: {len(issues)}")

# Apply fix
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

#After (verify)
remaining = df[df['UnitPrice'] != df['UnitPrice'].round(2)]
print("After fix:")
print(f"  UnitPrice precision issues remaining : {len(remaining)}")
print(f"  TotalPrice precision issues remaining: {len(df[df['TotalPrice'] != df['TotalPrice'].round(2)])}")
print(f"  Sample UnitPrice values: {df['UnitPrice'].head(5).tolist()}")
print("  STATUS: RESOLVED ")
```

```python
## Section 5:Standardize Date Format
#Issue:Date column is stored as a datetime object with no consistent string format
#Decision:Convert to ISO 8601 standard — YYYY-MM-DD

#Justification:ISO 8601 is the international standard for date representation.
#It ensures compatibility across systems, databases, and analytical tools.
```

```python
#Before
print("Before fix:")
print(f"  Data type : {df['Date'].dtype}")
print(f"  Sample    : {df['Date'].head(5).tolist()}")

#Apply fix
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')

#After (verify)
print("After fix:")
print(f"  Data type : {df['Date'].dtype}")
print(f"  Sample    : {df['Date'].head(5).tolist()}")

# Validate all dates match YYYY-MM-DD
import re
invalid_dates = df[~df['Date'].str.match(r'^\d{4}-\d{2}-\d{2}$')]
print(f"  Invalid date formats remaining: {len(invalid_dates)}")
print("  STATUS: RESOLVED")
```

```python
##Final Verification
#Running all verification checks required
#: 0% error rate on unique identifiers and date formats.
```

```python
print("=" * 50)
print("        FINAL VERIFICATION REPORT")
print("=" * 50)

checks = {
    "Duplicate OrderIDs = 0"         : df['OrderID'].duplicated().sum() == 0,
    "Null values in CouponCode = 0"  : df['CouponCode'].isnull().sum() == 0,
    "All dates in YYYY-MM-DD format" : df['Date'].str.match(r'^\d{4}-\d{2}-\d{2}$').all(),
    "UnitPrice precision correct"    : (df['UnitPrice'] == df['UnitPrice'].round(2)).all(),
    "TotalPrice precision correct"   : (df['TotalPrice'] == df['TotalPrice'].round(2)).all(),
    "Total row count preserved"      : len(df) == 1200,
}

all_passed = True
for check, result in checks.items():
    status = "PASS ✓" if result else "FAIL ✗"
    if not result:
        all_passed = False
    print(f"  {status}  |  {check}")

print("=" * 50)
print(f"  Overall: {'ALL CHECKS PASSED ✓' if all_passed else 'SOME CHECKS FAILED ✗'}")
print(f"  Final row count : {len(df)}")
print(f"  Final col count : {len(df.columns)}")
print("=" * 50)
```
